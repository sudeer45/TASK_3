```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data=pd.read_csv('c:\\Users\\LENOVO\\Downloads\\householdtask3.csv')
```

```python
display(data.head(25))
```

```python
#scatter plot with age against expenditure
plt.scatter (data['age'],data['expenditure'])
#adding tittle to the plot
plt.title("scatter plot")
#setting the x and y labels
plt.xlabel('age')
plt.ylabel('expenditure')
#adding the legends
plt.show()
```

```python
#line chart with age against expenditure
plt.plot(data['age'])
plt.plot(data['expenditure'])
#adding title
plt.title("line plot")
#setting the x and y labels
plt.xlabel('age')
plt.ylabel('expenditure')
#adding the legends
plt.show()
```

```python
#bar chart or bar plot
plt.bar(data['age'],data['expenditure'])
#adding tittle to the plot
plt.title("bar chart")
#setting the x and y labels
plt.xlabel('age')
plt.ylabel('expenditure')
#adding the legends
plt.show()
```

```python
#histogram
plt.hist(data['eqv_exp'])
plt.title('histogram')
plt.show()
```
